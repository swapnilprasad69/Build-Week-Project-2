```python
try:
    mydb = connection.connect(host="localhost",
                              user="root",
                              password="Vaishu@09",
                              database="quotes1",
                              use_pure=True)
    cursor = mydb.cursor()

    # 1. Number of quotes by each author
    query1 = """
        SELECT author, COUNT(DISTINCT quote) AS Number_of_Quotes
        FROM quotes_data
        GROUP BY author
    """
    df1 = pd.read_sql(query1, mydb)
    print("1. Quotes by each author:\n", df1, "\n")

    # 2. Top 5 most common tags
    query2 = """
        SELECT tag, COUNT(*) AS total
        FROM quotes_data
        GROUP BY tag
        ORDER BY total DESC
        LIMIT 5
    """
    df2 = pd.read_sql(query2, mydb)
    print("2. Top 5 tags:\n", df2, "\n")

    # 3. Authors with more than 5 quotes
    query3 = """
        SELECT author, COUNT(DISTINCT quote) AS Number_of_Quotes
        FROM quotes_data
        GROUP BY author
        HAVING Number_of_Quotes > 5
    """
    df3 = pd.read_sql(query3, mydb)
    print("3. Authors with >5 quotes:\n", df3, "\n")

    # 4. Longest quote and its author
    query4 = """
        SELECT author, quote, LENGTH(quote) AS Quote_Length
        FROM quotes_data
        ORDER BY Quote_Length DESC
        LIMIT 1
    """
    df4 = pd.read_sql(query4, mydb)
    print("4. Longest quote:\n", df4, "\n")

    # 5. Quotes that include "life"
    query5 = """
        SELECT DISTINCT quote
        FROM quotes_data
        WHERE quote LIKE '%life%'
    """
    df5 = pd.read_sql(query5, mydb)
    print("5. Quotes with 'life':\n", df5, "\n")

    # 6. Smallest quote
    query6 = """
        SELECT author, quote, LENGTH(quote) AS Quote_Length
        FROM quotes_data
        ORDER BY Quote_Length
        LIMIT 1
    """
    df6 = pd.read_sql(query6, mydb)
    print("6. Smallest quote:\n", df6, "\n")

    # 7. Average length of quotes
    query7 = """
        SELECT ROUND(AVG(LENGTH(quote)), 2) AS Avg_Quote_Length
        FROM quotes_data
    """
    df7 = pd.read_sql(query7, mydb)
    print("7. Average quote length:\n", df7, "\n")

except Exception as e:
    print(str(e))
finally:
    cursor.close()
    mydb.close()
```
